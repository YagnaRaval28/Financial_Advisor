```python
import pandas as pd
from fastapi import FastAPI, WebSocket, WebSocketDisconnect, HTTPException, Request
from fastapi.middleware.cors import CORSMiddleware
from pydantic import BaseModel
import nest_asyncio
from pyngrok import ngrok
import uvicorn
import os
import warnings
from typing import Optional, List
from langchain.docstore.document import Document
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_community.llms import HuggingFaceHub
from langchain.chains import RetrievalQA

app = FastAPI(title="AI Investment Advisor via WebSocket")

# Enable CORS
app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)

# Load API Key
os.environ["HUGGINGFACEHUB_API_TOKEN"] = "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"

# Initialize AI Model
warnings.filterwarnings('ignore')
qa_chain = None

# Load and Prepare Data
try:
    file_path = "drive/MyDrive/Finance_data.csv"
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"File '{file_path}' not found.")

    data = pd.read_csv(file_path)
    data_fin = data.to_dict(orient='records')

    prompt_response_data = []
    for entry in data_fin:
        prompt = (
            f"I'm a {entry['age']}-year-old {entry['gender']} looking to invest in "
            f"{entry['Avenue']} for {entry['Purpose']} over the next {entry['Duration']}. "
            f"What are my options?"
        )
        response = (
            f"Based on your preferences, here are your investment options:\n"
            f"- Mutual Funds: {entry['Mutual_Funds']}\n"
            f"- Equity Market: {entry['Equity_Market']}\n"
            f"- Debentures: {entry['Debentures']}\n"
            f"- Government Bonds: {entry['Government_Bonds']}\n"
            f"- Fixed Deposits: {entry['Fixed_Deposits']}\n"
            f"- PPF: {entry['PPF']}\n"
            f"- Gold: {entry['Gold']}\n"
        )
        prompt_response_data.append({"prompt": prompt, "response": response})

    documents = [
    Document(page_content=entry["response"])
    for entry in prompt_response_data
]

    text_splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=10)
    texts = text_splitter.split_documents(documents)

    hg_embeddings = HuggingFaceEmbeddings()
    persist_directory = 'docs/chroma/'
    vectordb_fin = Chroma.from_documents(
        documents=texts,
        embedding=hg_embeddings,
        persist_directory=persist_directory
    )

    print("🔍 Vector Store Ready")

    llm = HuggingFaceHub(
        repo_id="microsoft/Phi-3.5-mini-instruct",
        model_kwargs={"temperature": 0.1, "max_new_tokens": 500},
    )

    retriever_fin = vectordb_fin.as_retriever(search_kwargs={"k": 5})
    qa_chain = RetrievalQA.from_chain_type(
        llm=llm,
        chain_type="stuff",
        retriever=retriever_fin,
        return_source_documents=True
    )
    print("✅ AI Model Ready")

except Exception as e:
    print(f"❌ Error Initializing AI: {str(e)}")

# WebSocket Endpoint
@app.websocket("/ws/investment")
async def websocket_endpoint(websocket: WebSocket):
    await websocket.accept()
    try:
        while True:
            data = await websocket.receive_json()
            age = data.get("age")
            gender = data.get("gender")
            avenue = data.get("avenue")
            purpose = data.get("purpose")
            duration = data.get("duration")

            user_query = (
                f"I'm a {age}-year-old {gender} looking to invest in "
                f"{avenue} for {purpose} over the next {duration}. What are my options?"
            )

            result = qa_chain({"query": user_query})
            sources = [doc.page_content for doc in result.get("source_documents", [])]

            await websocket.send_json({
                "result": result.get("result", "No recommendation generated."),
                "source_documents": sources,
                "model_used": "mistralai/Mistral-7B-Instruct-v0.1",
                "status": "success"
            })

    except WebSocketDisconnect:
        print("❌ WebSocket Disconnected")
    except Exception as e:
        await websocket.send_json({"error": str(e)})

# Run FastAPI with ngrok
if __name__ == "__main__":
    try:
        ngrok_tunnel = ngrok.connect(8000)
        print("🌐 Public URL:", ngrok_tunnel.public_url)
        print("📜 API Docs:", ngrok_tunnel.public_url + "/docs")
        nest_asyncio.apply()
        uvicorn.run(app, host="0.0.0.0", port=8000, reload=False)
    except Exception as e:
        print(f"❌ Server Failed: {str(e)}")
```
